<h4> 2) Top participating countries

```python
top10_countries = athletes_df.Team.value_counts().sort_values(ascending = False).head(10)
top10_countries
```

Plot of above data

```python
a = np.array(top10_countries)
print(a)
b = top10_countries.index
print(b)
```

```python
plt.figure(figsize = (12,8))
x = b
y = a
cls = ['grey' , 'blue' , 'royalblue' ,'goldenrod' , 'black' , 'green' , 'aqua' , 'cornsilk' , 'red','lightgrey']
plt.xlabel('Countries',size =15,color = 'r')
plt.ylabel('No. of participnats',size =15,color = 'r')
plt.title('Top Participating Countries',size =25,color = 'Black',loc = 'center')
plt.bar(x,y,color = cls)
plt.show()
```
